```python
import os
import pandas as pd
```

```python
# Loading the dataset
src_path = '/content/drive/MyDrive/Train_For_CNN'
```

```python
classes = os.listdir(src_path)
classes
```

```python
Image_names_list=[]
for files in classes:
  Image_files_list = os.listdir(os.path.join(src_path,files))
  for i in Image_files_list:
    Image_names_list.append(i)
```

```python
import cv2    # Resizing all images to same size
def Resize_Image_Data(path_of_Image_files_to_resize,size,count):   ### herfe count is an integer image id used for naming resized images of each class

  image_resized_list = []
  for image_file in path_of_Image_files_to_resize:
    image = cv2.imread(image_file)
    #image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    if image is None:
      print('Wrong path:', image_file)
    else:
      image_resized_list.append(cv2.resize(image, size))
      count= count+1
      cv2.imwrite('/content/drive/MyDrive/Resized_images' + '/' + str(count) + '.jpg',cv2.resize(image, size))

  return image_resized_list   ### return the LIST of all resized images
#plt.imshow(image)
```

```python
import glob
path = '/content/drive/MyDrive/Train_For_CNN'
file_names_with_path_list=[]
Resized_image_list = []
count =0
for files in classes:
  count =count+1000
  file_names_with_path_list = glob.glob(path+'/'+files+'/*.jpg')
  Resized_image_list.append(Resize_Image_Data(file_names_with_path_list,(128,128),count))
```
